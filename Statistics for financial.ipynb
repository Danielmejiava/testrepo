```python
#random walk
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import math

cant = int(input('How many step do you want ?'))

def randomwalk1D(n):
    x, y = 0, 0

    # Generate the time points [1, 2, 3, ... , n]
    timepoints = np.arange(n + 1)
    #print(timepoints)
    positions = [y]
    #print(positions)

    directions = ("UP", "DOWN")
    #print(directions)
    for i in range(1, n + 1):

        # Randomly select either UP or DOWN
        step = random.choice(directions)
        #print(step)
        # Move the object up or down
        if step == "UP":
            y += 1
        elif step == "DOWN":
            y -= 1
        
        #print(step)

        # Keep track of the positions
        positions.append(y)

    return timepoints, positions
    
rw1 = np.array(randomwalk1D(cant))
#print(rw1)
rw1Dos = rw1.transpose()
rw1Dos = pd.DataFrame(rw1Dos)
rw1Dos = rw1Dos.iloc[:,-1] 

print("Caminata 1:\n", rw1Dos)
rw2 = np.array(randomwalk1D(cant))
rw2Dos = rw2.transpose()
rw2Dos = pd.DataFrame(rw2Dos)
rw2Dos = rw2Dos.iloc[:,-1] 
print("Caminata 2:\n", rw2Dos)
#print('que es eston Dos',rw2)
rw3 = np.array(randomwalk1D(cant))
rw3Dos = rw3.transpose()
rw3Dos = pd.DataFrame(rw3Dos)
rw3Dos = rw3Dos.iloc[:,-1]
print("Caminata 3:\n", rw3Dos)

rw4 = np.array(randomwalk1D(cant))
rw4Dos = rw4.transpose()
rw4Dos = pd.DataFrame(rw4Dos)
rw4Dos = rw4Dos.iloc[:,-1]
print("Caminata 4:\n", rw4Dos)

plt.plot(rw1[0], rw1[1], 'r-', label="rw1")
plt.plot(rw2[0], rw2[1], 'g-', label="rw2")
plt.plot(rw3[0], rw3[1], 'b-', label="rw3")
plt.plot(rw4[0], rw4[1], 'y-', label="rw4")

plt.show()

frecuencia = pd.concat([rw1Dos,rw2Dos,rw3Dos,rw4Dos], axis= 1, ignore_index=True) #axis is to create a dataframe horizantally and ignore_index= True is to delete de error 
frecuencia = pd.DataFrame(frecuencia)
frecuencia = frecuencia.set_axis(['Uno', 'Dos', 'Tres','Cuatro'], axis=1, inplace=False)
frecuencia = frecuencia.iloc[1: , :]
esperanza = cant*(2*(0.5)-1)
#Varianza = 4*cant*(0.5)*(1-0,5)

print("Esperanza: ", esperanza)
print(frecuencia)
```
